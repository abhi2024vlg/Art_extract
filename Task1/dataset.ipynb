# Combine Dataset

```python
# Vertical merge (append rows)
merged_wikiart_dataset = pd.concat([merged_train_dataset, merged_val_dataset], axis=0, ignore_index=True)

# Drop duplicates and NaNs
merged_wikiart_dataset = merged_wikiart_dataset.drop_duplicates().dropna()

merged_wikiart_dataset
```

```python
print("Number of unique artist class" , merged_wikiart_dataset["artist"].nunique())
print("Number of unique genre class" , merged_wikiart_dataset["genre"].nunique())
print("Number of unique style class" , merged_wikiart_dataset["style"].nunique())
```

## Since original Style dataset had 27 unique values and our resultant dataset has 16 unique values. We need to fix the indexing for further model training/evaluation  

```python
merged_wikiart_dataset["style"].value_counts()
```

### Creating a one to one mapping

```python
# Ensure all values are of the same type (e.g., integer or string) if needed:
merged_wikiart_dataset['style'] = merged_wikiart_dataset['style'].astype(int)

# Get the sorted unique values
unique_styles = sorted(merged_wikiart_dataset['style'].unique())

# Create a mapping dictionary: original style -> new integer (0 to len(unique_styles)-1)
mapping = {style: idx for idx, style in enumerate(unique_styles)}

print("Mapping:", mapping)
```

```python
merged_wikiart_dataset['style_mapped'] = merged_wikiart_dataset['style'].map(mapping)
merged_wikiart_dataset
```

```python
merged_wikiart_dataset = merged_wikiart_dataset.drop(columns=["style"])

# Rename columns
merged_wikiart_dataset.columns = ["file", "artist","genre","style"]

merged_wikiart_dataset
```

```python
merged_wikiart_dataset.to_csv("merged_wikiart_dataset.csv",index=False)
```
